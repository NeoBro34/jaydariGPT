```python
text = 'Hello world this is a simple example to show how tokenization works in NLP and a sentence'
print(text)
```

```python
tokens = text.split()
print(tokens)
```

```python
vocab = {}
current_id = 0

for token in tokens:
  if token not in vocab:
    vocab[token] = current_id
    current_id += 1

print(vocab)
```

```python
encoded = [vocab[token] for token in tokens]
print(encoded)
```

```python
reverse_vocab = {id: token for token, id in vocab.items()}
decoded = " ".join(reverse_vocab[id] for id in encoded)
print(decoded)
```

```python
# Embedding (bu so'zlarni mashina tushinadigan tilga raqamlarga aylantirib beruvchi jarayon)
# "hello" => [0.41, 0.22, -0.77, .......]
# "hi"    => [0.34, 0.43, 0,56, 0.87, .......]
```

```python
# Embedding vector => dense numerical array | Vector Database
# 0 => [0.41, 0.22, -0.77, .......]
# 1 => [0.34, 0.43, 0,56, 0.87, .......]
```

```python
import torch
import torch.nn as nn

embedding_dim = 8
vocab_size = len(vocab)

embedding_layer = nn.Embedding(vocab_size, embedding_dim)

input_ids = torch.tensor([encoded])
embeddings = embedding_layer(input_ids)

print(embeddings)
print(embeddings.shape)
```

```python
from transformers import AutoTokenizer

model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
print('Vocab size:', tokenizer.vocab_size)
print('Special tokens:', tokenizer.special_tokens_map)
```

```python
text = "Hello how are you doing?"
print('text:', text)
```

```python
tokens = tokenizer.tokenize(text)
print('tokens:', tokens)
```

```python
ids = tokenizer.convert_tokens_to_ids(tokens)
print('ids:', ids)
```

```python
encoded = tokenizer(text)
print('encoded:', encoded)
```

```python
print('Input IDs:', encoded['input_ids'])
```

```python
"Hi" # => [1, 345, 2, 2, 2]  // length 5
"How are you?" # => [1, 45, 657, 234, 2343] // length 5
# 'attention_mask': [1, 1, 0, 0, 0] -> bu tepada "Hi" length teng bo'lmagani uchun 1, 1 to'g'ri ma'lumot 0, 0, 0 fake malumot degan ma'noda sabab model to'g'ri training qilishi uchun lengthlar teng bo'lishi kerak
```

```python
decoded_text = tokenizer.decode(encoded['input_ids'])
print('decoded_text:', decoded_text)
```

```python
texts = [
    "Hello, how are you?",
    "Today we are learning about tokenizers.",
    "TinyLlama is a small but powerfull language model."
]

batch = tokenizer(texts, padding=True, return_tensors='pt')
print('Batch:', batch.keys())

print("Input IDs shape:", batch['input_ids'].shape)
print("Attention Mask shape:", batch['attention_mask'].shape)
```
